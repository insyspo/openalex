# Tables

BigQuery SQL codes to extract tables from the raw JSON tables as read from the data dump of OpenAlex.

Not all fields are covered, but nearly all of them, useful for nearly all searches. You can check against the documentation.

[OpenAlex entities](https://docs.openalex.org/api-entities/entities-overview)

## First level tables

### Authors

```python
%%bigquery --project=insyspo

DROP TABLE IF EXISTS insyspo.publicdb_openalex_2024_02_rm.authors;
CREATE TABLE insyspo.publicdb_openalex_2024_02_rm.authors AS (
  SELECT  TRIM(JSON_EXTRACT(author, '$.id'),'"') AS id,
    TRIM(JSON_EXTRACT(author, '$.display_name'),'"') AS display_name,
    TRIM(JSON_EXTRACT(author, '$.orcid'),'"') AS orcid,
    CAST(TRIM(JSON_EXTRACT(author, '$.cited_by_count'),'"') AS INT64) AS cited_by_count,
    TRIM(JSON_EXTRACT(author, '$.most_cited_work'),'"') AS most_cited_work,
    CAST(TRIM(JSON_EXTRACT(author, '$.created_date'),'"') AS DATETIME) AS created_date,
    CAST(TRIM(JSON_EXTRACT(author, '$.updated_date'),'"') AS DATETIME) AS updated_date,
    CAST(TRIM(JSON_EXTRACT(author, '$.works_count'),'"') AS INT64) AS works_count,
    CAST(TRIM(JSON_EXTRACT(author, '$.updated'),'"') AS DATETIME) AS updated,
    TRIM(JSON_EXTRACT(author, '$.last_known_institution'),'"') AS last_known_institution,
  FROM `insyspo.publicdb_openalex_2024_02.authors`
);
```

### Works

```python
%%bigquery --project=insyspo

DROP TABLE IF EXISTS insyspo.publicdb_openalex_2024_02_rm.works;
CREATE TABLE insyspo.publicdb_openalex_2024_02_rm.works AS (
  SELECT
    TRIM(JSON_EXTRACT(work, '$.id'),'"') AS id,
    CAST(TRIM(JSON_EXTRACT(work, '$.is_retracted'),'"') AS BOOL) AS is_retracted,
    TRIM(JSON_EXTRACT(work, '$.display_name'),'"') AS display_name,
    TRIM(JSON_EXTRACT(work, '$.title'),'"') AS title,
    CAST(TRIM(JSON_EXTRACT(work, '$.institutions_distinct_count'),'"') AS INT64) AS institutions_distinct_count,
    CAST(TRIM(JSON_EXTRACT(work, '$.locations_count'),'"') AS INT64) AS locations_count,
    CAST(TRIM(JSON_EXTRACT(work, '$.referenced_works_count'),'"') AS INT64) AS referenced_works_count,
    CAST(TRIM(JSON_EXTRACT(work, '$.authors_count'),'"') AS INT64) AS authors_count,
    CAST(TRIM(JSON_EXTRACT(work, '$.concepts_count'),'"') AS INT64) AS concepts_count,
    CAST(TRIM(JSON_EXTRACT(work, '$.cited_by_count'),'"') AS INT64) AS cited_by_count,
    TRIM(JSON_EXTRACT(work, '$.type'),'"') AS type,
    TRIM(JSON_EXTRACT(work, '$.cited_by_api_url'),'"') AS cited_by_api_url,
    CAST(TRIM(JSON_EXTRACT(work, '$.is_paratext'),'"') AS BOOL) AS is_paratext,
    CAST(TRIM(JSON_EXTRACT(work, '$.created_date'),'"') AS DATETIME) AS created_date,
    CAST(TRIM(JSON_EXTRACT(work, '$.publication_date'),'"') AS DATETIME) AS publication_date,
    COALESCE(
      CAST(SAFE.PARSE_TIMESTAMP('%FT%H:%M:%E*SZ',
              TRIM(JSON_EXTRACT(work, '$.updated'),'"')) AS DATETIME),
      CAST(TRIM(JSON_EXTRACT(work, '$.updated'),'"') AS DATETIME)
    ) AS updated,
    CAST(TRIM(JSON_EXTRACT(work, '$.updated_date'),'"') AS DATETIME) AS updated_date,
    CAST(TRIM(JSON_EXTRACT(work, '$.publication_year'),'"') AS INT64) AS publication_year,
    TRIM(JSON_EXTRACT(work, '$.doi'),'"') AS doi,
    TRIM(JSON_EXTRACT(work, '$.language'),'"') AS language,
    TRIM(JSON_EXTRACT(work, '$.open_access.oa_status'),'"') AS oa_status,
    TRIM(JSON_EXTRACT(work, '$.open_access.oa_url'),'"') AS oa_url,
    CAST(TRIM(JSON_EXTRACT(work, '$.open_access.is_oa'),'"') AS BOOL) AS is_oa,
    CAST(TRIM(JSON_EXTRACT(work, '$.open_access.any_repository_has_full_text'),'"') AS BOOL) AS any_repository_has_full_text,
 FROM `insyspo.publicdb_openalex_2024_02.works`
);
```

## Second level tables

### Publishers auxiliary tables

#### Publishers lineage


```python
%%bigquery --project=insyspo

DROP TABLE IF EXISTS insyspo.publicdb_openalex_2024_02_rm.publishers_lineages;
CREATE TABLE insyspo.publicdb_openalex_2024_02_rm.publishers_lineages AS (
  WITH aux AS (
    SELECT
      TRIM(JSON_EXTRACT(publisher, '$.id'),'"') AS publisher_id,
      JSON_EXTRACT_ARRAY(publisher, '$.lineage') AS lineages,
    FROM `insyspo.publicdb_openalex_2024_02.publishers`
    WHERE ARRAY_LENGTH(JSON_EXTRACT_ARRAY(publisher,'$.lineage'))>0
  )
  SELECT publisher_id,
    TRIM(JSON_EXTRACT(lineage, '$.lineage'),'"') AS lineage
  FROM aux,
  UNNEST(lineages) lineage
);
```

####Publishers concepts

```python
%%bigquery --project=insyspo

DROP TABLE IF EXISTS insyspo.publicdb_openalex_2024_02_rm.publishers_concepts;
CREATE TABLE insyspo.publicdb_openalex_2024_02_rm.publishers_concepts AS (
  WITH aux AS (
    SELECT
      TRIM(JSON_EXTRACT(publisher, '$.id'),'"') AS publisher_id,
      JSON_EXTRACT_ARRAY(publisher,'$.x_concepts') AS concepts
    FROM `insyspo.publicdb_openalex_2024_02.publishers`
    WHERE ARRAY_LENGTH(JSON_EXTRACT_ARRAY(publisher,'$.x_concepts'))>0
  )
  SELECT
      publisher_id,
      CAST(TRIM(JSON_EXTRACT(concept, '$.score'),'"') AS FLOAT64) AS score,
      TRIM(JSON_EXTRACT(concept, '$.id'),'"') AS concept_id,
  FROM aux,
  UNNEST(concepts) concept
);
```

### Concepts auxiliary tables

#### Concepts related concepts



```python
%%bigquery --project=insyspo

DROP TABLE IF EXISTS insyspo.publicdb_openalex_2024_02_rm.concepts_related_concepts;
CREATE TABLE insyspo.publicdb_openalex_2024_02_rm.concepts_related_concepts AS (
  WITH aux AS (
    SELECT  TRIM(JSON_EXTRACT(concept, '$.id'),'"') AS id,
      JSON_EXTRACT_ARRAY(concept, '$.related_concepts') AS related_concepts,
    FROM `insyspo.publicdb_openalex_2024_02.concepts`
    WHERE ARRAY_LENGTH(JSON_EXTRACT_ARRAY(concept, '$.related_concepts'))>0
  )
  SELECT id AS concept_id,
    CAST(TRIM(JSON_EXTRACT(related_concept, '$.score'),'"') AS FLOAT64) AS score,
    TRIM(JSON_EXTRACT(related_concept, '$.id'),'"') AS related_concept_id,
  FROM aux,
  UNNEST(related_concepts) related_concept
)
```

#### Concepts ancestors

```python
%%bigquery --project=insyspo

DROP TABLE IF EXISTS insyspo.publicdb_openalex_2024_02_rm.concepts_ancestors;
CREATE TABLE insyspo.publicdb_openalex_2024_02_rm.concepts_ancestors AS (
   WITH aux AS (
    SELECT  TRIM(JSON_EXTRACT(concept, '$.id'),'"') AS id,
      JSON_EXTRACT_ARRAY(concept, '$.ancestors') AS concepts_ancestors,
    FROM `insyspo.publicdb_openalex_2024_02.concepts`
    WHERE ARRAY_LENGTH(JSON_EXTRACT_ARRAY(concept, '$.ancestors'))>0
  )
  SELECT id AS concept_id,
    TRIM(JSON_EXTRACT(concept_ancestor, '$.id'),'"') AS ancestor_concept_id,
  FROM aux,
  UNNEST(concepts_ancestors) concept_ancestor
)
```

# Abstracts from inverted abstracts

A Javascript UDF is created to build the abstracts from the inverted index. It is created on a different dataset.

```python
%%bigquery --project=insyspo

DROP FUNCTION IF EXISTS insyspo.projectdb_cwts_summer_school_2023.read_join_json;
CREATE FUNCTION insyspo.projectdb_cwts_summer_school_2023.read_join_json(d STRING)
RETURNS STRING
LANGUAGE js
AS r"""
  try {
    obj = JSON.parse(d);
    var arr = [];
    Object.keys(obj.InvertedIndex).forEach(k => obj.InvertedIndex[k].forEach(a => arr[a] = k));
    text = arr.join(' ');
  } catch {
    text = ''
  }

  return text;
""";

DROP TABLE IF EXISTS insyspo.projectdb_cwts_summer_school_2023.works_abstracts;
CREATE TABLE insyspo.projectdb_cwts_summer_school_2023.works_abstracts AS (
  SELECT id,
    insyspo.projectdb_cwts_summer_school_2023.read_join_json(abstract_inverted_index) AS abstract
  FROM `insyspo.publicdb_openalex_2024_02_rm.works_inverted_abstracts`
  WHERE abstract_inverted_index IS NOT NULL
)
```
